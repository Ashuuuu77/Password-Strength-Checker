Initial version (strength check)

```python
def check_strength(password):
    score = 0

    if len(password) >= 8:
        score += 1
    if any(char.isdigit() for char in password):
        score += 1
    if any(char.isupper() for char in password):
        score += 1
    if any(char in "!@#$%^&*" for char in password):
        score += 1

    if score <= 2:
        return "Weak Password"
    elif score == 3:
        return "Moderate Password"
    else:
        return "Strong Password"

pwd = input("Enter password: ")
print(check_strength(pwd))
```

Common password detection

```python
common_passwords = ["123456", "password", "qwerty", "abc123"]

def is_common(password):
    return password in common_passwords

pwd = input("Enter password again: ")

if is_common(pwd):
    print("⚠️ This is a commonly used password!")
else:
    print("Not a common password")
```

Final combined version

```python
def check_strength(password):
    score = 0

    if len(password) >= 8:
        score += 1
    if any(char.isdigit() for char in password):
        score += 1
    if any(char.isupper() for char in password):
        score += 1
    if any(char in "!@#$%^&*" for char in password):
        score += 1

    if score <= 2:
        return "Weak Password"
    elif score == 3:
        return "Moderate Password"
    else:
        return "Strong Password"


common_passwords = ["123456", "password", "qwerty", "abc123"]

def is_common(password):
    return password in common_passwords


pwd = input("Enter password: ")

print("Strength:", check_strength(pwd))

if is_common(pwd):
    print("⚠️ Common password detected!")
```
